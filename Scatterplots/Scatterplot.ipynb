```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset("exercise")
df.head()
```

```python
# Scatterplot

sns.scatterplot(x = "time", y = "pulse", data = df)

```

```python
# hue 

sns.scatterplot(x = "time", y = "pulse", data = df, hue = "kind")
```

```python

sns.scatterplot(x = "time", y = "pulse", data = df, hue = "kind", palette = "GnBu")
```

```python
# Marker size depending upon diet

sns.scatterplot(x = "time", y = "pulse", data = df, hue = "kind", size = "diet")
```

```python
# Marker shape

sns.scatterplot(x = "time", y = "pulse", data = df, marker = "s") # "*,^"
plt.show()
```

```python
# To change transparency alpha = 0-1

sns.scatterplot(x = "time", y = "pulse", data = df, hue = "diet", alpha = 0.7)
plt.legend()
plt.show()
```

```python
sns.scatterplot( x = 'diet', y = 'pulse', data = df, hue = 'time')
plt.show()
```
