```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)

sns.kdeplot(x)      # single axis "x"
```

```python
sns.kdeplot(y)  # Single axis "y"
```

```python
sns.kdeplot(y, shade = True) # To shade in the area under graph
```

```python
sns.kdeplot(y, shade = True, color = "violet")  #to change color
```

```python
# multiple columns/axis

sns.kdeplot(x, y)
```

```python
# multiple columns/axis

sns.kdeplot(x=x, y=y)
```

```python
# multiple columns/axis

sns.kdeplot(x=x, y=y, shade = True) 
```

```python
sns.kdeplot(x=x, y=y, shade = True, cmap = "autumn")  # For "palette" we use "cmap"
```

```python
sns.kdeplot(x=x, y=y, shade = True, cmap = "GnBu") 
```
