```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

x = [1,2,3,4,5,6,7]
y = [20,58,35,85,26,86,64]

df = pd.DataFrame({"Days: ": x, "Number of People: ": y})
df.head(7)
```

```python
# Seaborn plot by creating DF

sns.lineplot(x = "Days: ", y = "Number of People: ", data = df)
plt.show()
```

```python
# Seaborn plot by loading DF

df = pd.read_csv(r"C:\Users\ayush\Desktop\Projects -Data ANalyst\Netflix\netflix_titles.csv")
df.head()
```

```python
# Step 1: Group by 'release_year' and 'type', count entries
type_counts = df.groupby(['release_year', 'type']).size().reset_index(name='count')

# Step 2: Plot
sns.lineplot(x="release_year", y="count", hue="type", data=type_counts, marker="o")
plt.title("Count of Movies and TV Shows per Year")
plt.ylabel("Count")
plt.xlabel("Release Year")
plt.show()
```

```python
type_counts = df.groupby(['release_year', 'type']).size().reset_index(name='count')

# Step 2: Plot
sns.lineplot(x="release_year", y="count", hue="type", data=type_counts)
plt.title("Count of Movies and TV Shows per Year")
plt.ylabel("Count")
plt.xlabel("Release Year")
plt.show()
```

```python
type_counts = df.groupby(['release_year', 'type', 'rating']).size().reset_index(name='count')

# Step 2: Plot
sns.lineplot(x="release_year", y="count", hue="rating", data=type_counts, marker="o")
plt.title("Count of Movies and TV Shows per Year by rating")
plt.ylabel("Count")
plt.xlabel("Release Year")
plt.tight_layout()
plt.show()
```

```python
g = sns.FacetGrid(type_counts, col="type", hue="rating", height=5, aspect=1.5)
g.map_dataframe(sns.lineplot, x="release_year", y="count", marker="o")
g.add_legend()
g.set_axis_labels("Release Year", "Count")
g.fig.suptitle("Shows per Year by Type and Rating", y=1.03)
plt.show()
```
