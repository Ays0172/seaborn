```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset('tips')
df.head()
```

```python
sns.stripplot(x = df['tip'], data = df)
```

```python
# btw x and y axis

sns.stripplot(x = 'day', y = 'tip', data = df)
```

```python
sns.stripplot(x = 'day', y = 'tip', data = df, hue = 'sex')
```

```python
# Jitter
sns.stripplot(x = 'day', y = 'tip', data = df, palette = 'GnBu')  # without jitter
```

```python
# Jitter
sns.stripplot(x = 'day', y = 'tip', data = df, palette = 'GnBu', jitter = 0.4)  # with jitter -  to identify the no of elements in an area
```

```python
sns.stripplot(x = 'day', y = 'tip', data = df, palette = 'GnBu', jitter = 0.4, hue = 'sex')
```

```python
#Dodge

sns.stripplot(x = 'day', y = 'tip', data = df, palette = 'GnBu', jitter = 0.4, hue = 'sex', dodge=True) # Boolean values, seperated categories
```

```python
sns.stripplot(x = 'day', y = 'tip', data = df, marker = '*', size = 10, alpha = 0.8)
```

```python
sns.stripplot(x = 'day', y = 'tip', data = df, marker = '*', size = 10, alpha = 0.8, palette = 'GnBu', jitter = 0.3)
```

```python
# Linewidth

sns.stripplot(x = 'day', y = 'tip', data = df, marker = '*', size = 10, alpha = 0.8, palette = 'GnBu', jitter = 0.3, linewidth = 1, edgecolor = 'black')
```
