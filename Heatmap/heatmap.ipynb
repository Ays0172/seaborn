```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randint(low = 1, high = 100, size = (10,10))
print(data)
```

```python
# For heatmaps

sns.heatmap(data = data)
```

```python
# color maps

sns.heatmap(data = data, cmap = 'GnBu')
```

```python
# transperancy

sns.heatmap(data = data, cmap = 'GnBu', alpha = 0.8)
```

```python
# Line color, Linewidth

sns.heatmap(data = data, cmap = 'GnBu', linecolor = 'yellow', linewidth = '2')
```

```python
# To remove colorbar and to see the values

sns.heatmap(data = data, cbar = False)  # Colorbar
```

```python
# Annotations
# transperancy

sns.heatmap(data = data, annot = True)
```

```python
# Ticklabels

sns.heatmap(data = data, xticklabels = False, yticklabels = False, linecolor = 'blue', linewidth = '1')
```
