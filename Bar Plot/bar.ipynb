```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df =sns.load_dataset("titanic")
df.head()
print(df)
```

```python
sns.barplot(x = "who", y = "fare", hue = "who", data = df)
plt.legend()
plt.show()
```

```python
sns.barplot(x = "who", y = "fare", hue = "who", data = df, color = "grey")
plt.show()
```

```python
sns.barplot(x = "who", y = "fare", hue = "sex", data = df)
plt.show()
```

```python
sns.barplot(x = "sex", y = "fare", hue = "who", data = df)
plt.show()
```

```python
sns.barplot(x = "sex", y = "fare", hue = "who", data = df, palette = "spring")
plt.show()
```

```python
sns.barplot(x = "sex", y = "fare", hue = "who", data = df, palette = "spring", saturation = 0.2) # to change the intensity of colors
plt.show()
```

```python
# For Horizontal Bar
sns.barplot(x = "fare", y = "who", hue = "who", data = df)
plt.show()
```

```python
# To remove median lines

sns.barplot(x = "who", y = "fare", hue = "who", data = df, ci = False)  #Using ci function
plt.show()
```

```python
# To see mean instead of median

from numpy import mean
sns.barplot(x = "fare", y = "who", hue = "who", data = df, estimator = mean)  #Using estimator for mean
plt.show()

from numpy import median
sns.barplot(x = "fare", y = "who", hue = "who", data = df, estimator = median)  #Using estimator for median
plt.show()
```

```python
# To have plain white background

sns.barplot(x = "who", y = "fare", hue = "who", data = df, facecolor = [1,1,1,0], edgecolor = "0.5")  # [r,g,b,alpha]
plt.show()
```

```python
# To change order in which they appear - child, women, man

sns.barplot(x = "who", y = "fare", hue = "who", data = df, order = ["child", "woman", "man"])  # order
plt.show()
```
