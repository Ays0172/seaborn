```python
# It counts the value and gives the value of thosw

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')
print(df)
```

```python
#Countplot

sns.countplot( x = 'tip', data = df)
```

```python
sns.countplot(x='sex', data=df, hue = 'sex')
```

```python
# define "y - axis" for horizontal bar

sns.countplot(y='sex', data=df, hue = 'sex')
```

```python
# hue for sub catogrisation

sns.countplot(x='sex', data=df, hue = 'smoker', palette = 'autumn')
```

```python
# For same color

sns.countplot(x='sex', data=df, color = 'green')
```

```python


sns.countplot( x = 'sex', data =df, facecolor = [0,0,0,0,], linewidth = 4, edgecolor = sns.color_palette('GnBu', 2))
```
