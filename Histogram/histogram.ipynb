```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.histplot(x = "total_bill", data = df)
plt.show()
```

```python
# To give gaps btw bars

sns.histplot(x = "total_bill", data = df, discrete = "false")       # False - Treats your data as continuous
plt.show()
```

```python
# To give gaps btw bars

sns.histplot(x = "total_bill", data = df, discrete = "true")        # true - Treats your data as discrete
plt.show()
```

```python
# Hue - overlapping data

sns.histplot(x = "total_bill", data = df, hue = "sex")        
plt.show()
```

```python
# Hue - distinct palette

sns.histplot(x = "total_bill", data = df, hue = "sex", palette = "spring")        
plt.show()
```

```python
# To Apply kernel density 

sns.histplot(x = "total_bill", data = df, kde = True) # Kernel Density estimation
plt.show()
```

```python
# To Apply Edge color

sns.histplot(x = "total_bill", data = df, kde = True, edgecolor = "red") # Edge color
plt.show()
```

```python
# To Apply color

sns.histplot(x = "total_bill", data = df, kde = True, color = "hotpink") #  color
plt.show()
```
