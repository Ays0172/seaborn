```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset('tips')
df.head()
```

```python
# For single violin plot

sns.violinplot(x = df['total_bill'])
```

```python
# For x and y axis

sns.violinplot(x = 'total_bill', y = "tip", data = df )
```

```python
# For x and y axis

sns.violinplot(x = 'sex', y = "tip", data = df, palette = 'autumn')
```

```python


sns.violinplot(x = 'sex', y = "tip", data = df, hue = 'day')
```

```python
# Vertical plot

sns.violinplot(y = 'sex', x = "tip", data = df, hue = 'day')
```

```python
# Split function

sns.violinplot(y = 'tip', x = "day", data = df, hue = 'sex', split = True)
```

```python
# Saturation

sns.violinplot(y = 'tip', x = "day", data = df, hue = 'sex', saturation = 0.3)
```

```python
# To change order to Female and Male plots

sns.violinplot(x = 'sex', y = "tip", data = df, hue = 'day', order = ['Female', 'Male'])
```

```python
# Width of the lines

sns.violinplot(x = 'sex', y = "tip", data = df, hue = 'day', order = ['Female', 'Male'], linewidth = 2)
```
