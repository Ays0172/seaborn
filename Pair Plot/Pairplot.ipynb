```python
# Pair Plot - a combination of a Histograms and ScatterPlots
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset('iris')
df.head()
```

```python
sns.pairplot(data = df)
```

```python
# To reduce the height

sns.pairplot(data = df, height = 1.5)
```

```python
# Hue

sns.pairplot(data = df, height = 1.5, hue = 'species')
```

```python
# Paleete

sns.pairplot(data = df, height = 1.5, hue = 'species', palette = 'viridis')
```

```python
# kde

sns.pairplot(data = df, height = 1.5, hue = 'species', palette = 'viridis', diag_kind = 'hist')
```

```python
# Edgecolor

sns.pairplot(data = df, height = 1.5, plot_kws = {"edgecolor" : "red"})
```
