```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset('exercise')
df.head()
```

```python
# FacetGrids - Multiple Grids

sns.FacetGrid(df, col = 'kind') #Only provided no of columns, but data not mentioned--so no data!
```

```python
a = sns.FacetGrid(df, col = 'kind')

a.map(sns.scatterplot, "time", 'pulse') # ("x" and "y" variables)
a.add_legend()
```

```python
a = sns.FacetGrid(df, col = 'kind', hue = "time")

a.map(sns.scatterplot, "time", 'pulse') # ("x" and "y" variables)
a.add_legend()
```

```python
# For bar plot
a = sns.FacetGrid(df, col = 'kind', hue = "time")
a.map(sns.barplot, "time", "pulse")
plt.show()
```

```python
# Specifying Order to avoid confusion!
a = sns.FacetGrid(df, col = 'kind', hue = "time")
a.map(sns.barplot, "time", "pulse", order = ["30 min", "15 min", "1 min"])
a.add_legend()
plt.show()
```

```python
# Using kde plot

a = sns.FacetGrid(df, col = 'kind', hue = "time")
a.map(sns.kdeplot, "pulse")
a.add_legend()
plt.show()
```

```python
a = sns.FacetGrid(df, col = 'diet')
a.map(sns.kdeplot, "pulse")
a.add_legend()
plt.show()
```

```python
# PairPlot

p = sns.PairGrid(df)
p.map_diag(sns.histplot)
p.map_offdiag(sns.scatterplot)
```

```python
# with hue

p = sns.PairGrid(df, hue="diet")
p.map_diag(sns.histplot)
p.map_offdiag(sns.scatterplot)
p.add_legend()
```

```python

p = sns.PairGrid(df, hue="time")
p.map_diag(sns.kdeplot)
p.map_offdiag(sns.scatterplot)
p.add_legend()
```
