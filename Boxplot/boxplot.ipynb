```python
import seaborn as sns
import matplotlib.pyplot as mlt
import pandas as pd

df = sns.load_dataset('tips')
df.head()
```

```python
# Boxplot
#single variable (x)

sns.boxplot( x = df['tip'], data = df )
```

```python
#Multiple variables

sns.boxplot( x = 'tip', y = 'sex', data = df)
```

```python
sns.boxplot( x = 'sex', y = 'tip', data = df, palette = 'spring')
```

```python
sns.boxplot( x = 'sex', y = 'tip', data = df, hue = 'smoker')
```

```python
sns.boxplot( x = 'sex', y = 'tip', data = df, hue = 'smoker', palette = 'viridis')
```

```python
sns.boxplot( x = 'sex', y = 'tip', data = df, hue = 'smoker', palette = 'viridis', linewidth = 3)
```

```python
sns.boxplot(data = df, orient = "h")
```

```python
sns.boxplot(data = df, orient = "v")
```

```python
sns.set(style = "whitegrid")
sns.boxplot(data = df, orient = "v")
```
