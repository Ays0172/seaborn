```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset('tips')
df.head()
```

```python
sns.catplot(x = 'sex', y = 'tip', data = df, palette = 'spring')
# default value = scatterplot
```

```python
# To have bar chart
sns.catplot(x = 'sex', y = 'tip', data = df, palette = 'spring', kind = 'bar')
```

```python
# Violin plot

sns.catplot(x = 'sex', y = 'tip', data = df, palette = 'spring', kind = 'violin')
```

```python
# Violin plot

sns.catplot(x = 'sex', y = 'tip', data = df, palette = 'GnBu', kind = 'violin', hue = 'smoker')
```

```python
# Columns
sns.catplot(x = 'sex', y = 'tip', data = df, palette = 'GnBu', kind = 'violin', hue = 'smoker', col = 'time')
```

```python
# Height

sns.catplot(x = 'sex', y = 'tip', data = df, palette = 'GnBu', kind = 'violin', hue = 'smoker', col = 'time', height = 4)
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.catplot(x = 'sex', y = 'pclass', data = df, col = 'deck')
```

```python
sns.catplot(x = 'class', data = df, col = 'deck', col_wrap = 3) # 3 columns per row
```

```python
sns.catplot(x = 'class', data = df, col = 'deck', col_wrap = 3, height = 4, hue = "embarked") # 3 columns per row
```
