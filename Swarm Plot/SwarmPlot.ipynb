```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# They are different from Strip plots as they are more defined

df = sns.load_dataset('tips')
df.head()
```

```python
sns.swarmplot( x = 'day', y = 'tip', data = df) #better defined compared to strip plot
```

```python
sns.swarmplot( x = 'day', y = 'tip', data = df, size = 3)   # To reduce the size
```

```python
sns.swarmplot( x = 'day', y = 'total_bill', data = df, size = 3, palette = 'spring')   # To reduce the size
```

```python
sns.swarmplot( x = 'day', y = 'total_bill', data = df, palette = 'spring')   
```

```python
sns.swarmplot( x = 'day', y = 'total_bill', data = df, hue = "time")
```

```python
sns.swarmplot( x = 'day', y = 'total_bill', data = df, palette = 'spring', hue = 'time')
```

```python
sns.swarmplot( x = 'day', y = 'total_bill', data = df, hue = "smoker")
```

```python
sns.swarmplot( x = 'day', y = 'total_bill', data = df, hue = "smoker", dodge= True) # Two seperate smoker category further
```

```python
sns.stripplot( x = 'day', y = 'total_bill', data = df)      # For comparison of both plots
```

```python
sns.swarmplot( x = 'day', y = 'total_bill', data = df)
```

```python
sns.stripplot( x = 'day', y = 'total_bill', data = df, hue = "smoker", dodge= True) # Two seperate smoker category further
```

```python
sns.swarmplot( x = 'day', y = 'total_bill', data = df, hue = "smoker", dodge= True, marker = '*', order = ['Sun', 'Thur', 'Sat', 'Fri'], size = 8, edgecolor = 'gold', linewidth = 0.5)
```
